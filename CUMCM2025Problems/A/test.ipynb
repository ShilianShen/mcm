```python
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import parallel_coordinates
import pandas as pd

# 生成示例数据：5维向量，最后一维是优化目标
np.random.seed(42)
data = np.random.rand(20, 5)  # 20个5维向量
data[:, 4] = data[:, 0] + data[:, 1] * 2 - data[:, 2]  # 目标维度

# 创建DataFrame
df = pd.DataFrame(data, columns=['Dim1', 'Dim2', 'Dim3', 'Dim4', 'Target'])

# 平行坐标图
plt.figure(figsize=(10, 6))
parallel_coordinates(df, 'Target', color=('#556270', '#4ECDC4', '#C7F464'))
plt.title('5维向量平行坐标图（颜色表示目标值）')
plt.show()
```
